As data analyst, I need to explore the data and use machine learning to build a model to predict whether a customer is likely to buy.

# <font color="red">Part 2 - the Model</font>

Building the model.

#### Transformations required
    
- dummmy (one-hot coded) for categorical columns

<span style="color: red;"> `Target Column = y, Feature Columns = age, job, education, poutcome, campaign, and cons price index`

I need to dummify education, job, and poutcome </span>

```python
education_dummies = pd.get_dummies(bank.education,drop_first=True, dtype = int, prefix = 'education')
job_dummies = pd.get_dummies(bank.job,drop_first=True, dtype = int, prefix = 'job')
outcome_dummies = pd.get_dummies(bank.poutcome,drop_first=True, dtype = int, prefix = 'outcome')
```

#### Creating `X_train`, `X_test`, `y_train` and `y_test` variables using the features choden above

```python
X = pd.concat([education_dummies, job_dummies, outcome_dummies, bank['age'], bank['cons.price.idx'], bank['campaign']], axis=1)
y = bank.y
```

```python
# Using KNeighbors Classifier, simple
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=100)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred_class = knn.predict(X_test)
print(f"Accuracy Score is: {metrics.accuracy_score(y_test, y_pred_class)}")
print(f"Recall Score is: {metrics.recall_score(y_test, y_pred_class)}")
```

```python
# Using KNeighbors Classifier with Stratify and Standard Scaler

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 102, stratify = y)
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(X_train_sc, y_train)
preds = kn.predict(X_test_sc)

print(f"Accuracy Score is: {accuracy_score(y_test, preds)}")

print(f"Recall Score is: {recall_score(y_test, preds)}")
```

```python
# Using Smote as dataset is imbalanced, as explained below

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 102, stratify = y)
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
kn = KNeighborsClassifier(n_neighbors = 5)
kn.fit(X_train_sc, y_train)
preds = kn.predict(X_test_sc)

print(f"Accuracy Score is: {accuracy_score(y_test, preds)}")

print(f"Recall Score is: {recall_score(y_test, preds)}")
```

#### Evaluation of the model
    
Comparison of accuracy scores with the baseline.The baseline is "predict everything is the majority class" (in this case always predict no purchase). Accuracy score should be higher than the % of the majority class.

#### Next, confusion matrix


```python
# Using Smote for the confusion matrix because the recall score is the highest

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 102, stratify = y)
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
kn = KNeighborsClassifier(n_neighbors = 5)
kn.fit(X_train_sc, y_train)
preds = kn.predict(X_test_sc)

cm = confusion_matrix(y_test, preds)
cm

disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.show();
```

<span style="color: red;">In this model, we have more false positives than false negatives. A false positive is when we predicted a client would buy, but they didn't. The cost with this might be all of the wasted marketing campaigns money that the company spent on. A false negative is when we predicted a client would not buy, but they did. This represents a failed opportunity to engage with the client. </span>
